```python
Homeless_Ireland = pd.read_csv('Homeless_Ireland_2019_to_2021_.csv',index_col = 0)
Homeless_Ireland
```

```python
Homeless_Ireland = Homeless_Ireland.loc[:, ~Homeless_Ireland.columns.str.contains('^Unnamed')]
Homeless_Ireland
```

```python
fig, ax = plt.subplots()
sns.set_palette("RdBu")
Homeless_Ireland.plot()
ax.bar(Homeless_Ireland.index,Homeless_Ireland["Total Adults"])
ax.set_xlabel("Regions")
ax.set_ylabel("Total Adults")
plt.show()
```

```python
fig, ax = plt.subplots()
Homeless_Ireland.plot()
ax.plot(Homeless_Ireland.index,Homeless_Ireland["Total Adults"])
ax.scatter(Homeless_Ireland.index,Homeless_Ireland["Total Adults"],color = 'g')
ax.set_xlabel ("Regions")
ax.set_ylabel ("Total Adults (Number)")
plt.show()
```

```python
plt.style.use("ggplot")
fig, ax = plt.subplots()
Homeless_Ireland.plot()
ax.plot(Homeless_Ireland ["Female Adults"],Homeless_Ireland["Year"], color = "r")
ax.plot(Homeless_Ireland ["Male Adults"],Homeless_Ireland["Year"] , color = "g")
ax.set_xlabel ("Total Adults")
ax.set_ylabel ("Duration")
plt.show()
```
